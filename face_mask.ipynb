```python
import os
import cv2
```

```python
from keras.preprocessing import image
```

```python
categories = ['with_mask', "without_mask"]
```

```python
data = []

for category in categories:
    path = os.path.join('train', category)

    label = categories.index(category)

    for file in os.listdir(path):
        img_path = os.path.join(path,file)
        img = cv2.imread(img_path)
        img = cv2.resize(img,(224,224))
        print(img.shape)

        data.append([img,label])


        
        # print(file)
        # break
```

```python
import random
```

```python
random.shuffle(data)
```

```python
x = []
y = []

for feature, label in data:
    x.append(feature)
    y.append(label)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape
```

```python
from tensorflow.keras.applications import VGG19
```

```python
vgg = VGG19()
```

```python
vgg.summary()
```

```python
from keras import Sequential
```

```python
model = Sequential()
```

```python
for layer in vgg.layers[:-1]:
    model.add(layer)
```

```python
model.summary()
```

```python
for layer in model.layers:
    layer.trainable = False
```

```python
model.summary()
```

```python
from keras.layers import Dense
```

```python
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()  #transfer learning
```

```python
model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=5, validation_data=(x_test,y_test))
```

```python
def draw_label(img,text,pos,bg_color):
    text_size = cv2.getTextSize(text,cv2.FONT_HERSHEY_SIMPLEX,1,cv2.FILLED)

    end_x = pos[0] + text_size[0][0] +2
    end_y = pos[1] + text_size[0][1] +2

    cv2.rectangle(img,pos,(end_x,end_x),bg_color,cv2.FILLED)
    cv2.putText(img,text,pos,cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,0),1,cv2.LINE_AA)
```

```python
import cv2
```

```python
haar = cv2.CascadeClassifier('haarcascade_frontalcatface_extended.xml')
```

```python
def detect_face(img):
    coods = haar.detectMultiScale(img)
    # print(coods)
    return coods
```

```python
cap = cv2.VideoCapture(0) # 0 for webcam
```

```python
while True:
    ret, frame = cap.read()
    img = cv2.resize(frame,(224,224))
    y_pred = detect_face_mask(img)
    
    coods = detect_face(cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY))
    for x,y,w,h in coods:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),1)

    if y_pred == 0:
        draw_label(frame, "Mask", (30,30), (0,225,0))
    else:
        draw_label(frame, "No Mask", (30, 30), (0, 0, 225))
        
    cv2.imshow("window",frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cv2.destroyAllWindows()
```
